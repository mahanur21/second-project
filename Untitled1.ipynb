```python
x=('/content/drive/My Drive/ecommerse.csv')
import pandas as pd
df=pd.read_csv(x)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Reached.on.Time_Y.N', data=df)
plt.xlabel('Reached on Time (0 = Yes, 1 = No)')
plt.title('Delivery Timeliness Count')
plt.show()


```

```python
plt.scatter(df['Weight_in_gms'], df['Cost_of_the_Product'], c='green')
plt.title('Product Weight vs. Cost')
plt.xlabel('Weight (gms)')
plt.ylabel('Cost')
plt.grid(True)
plt.show()
```

```python
df['Product_importance'].value_counts().plot.pie(autopct='%1.1f%%', startangle=140)
plt.title('Product Importance Distribution')
plt.ylabel('')
plt.show()
```
